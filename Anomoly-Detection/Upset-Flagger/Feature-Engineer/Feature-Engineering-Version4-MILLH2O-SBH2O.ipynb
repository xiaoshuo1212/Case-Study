```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    dataframe: should be a (nrows x 2 column) size dataframe, with sum of column = 100 for each row
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax)
    # Rename legend
    plt.legend(["Normal", "Upset"], loc=legend_)
    # Labels
    plt.ylabel("Cases")
    plt.savefig('Target_Output/{}/{}.png'.format(area, title_))
    plt.show()
    

def annotate_stacked_bars(ax, pad=0.99, colour="black", textsize=10):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        # Calculate annotation
        value = str(p.get_height())
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

**Feature Correlation Analysis:**

- Ideally, for __parametric model__, we want a set of features that have 0 correlation with all of the independent variables (all features except our target variable) and a high correlation with the target variable.
- First, we measure the correlations between features and target, this is important because we want features to correlate with our target(upset or not), as this will indicate that the features are good predictors of it. From the correlation heatmap, we can see that the P86_Froth_Water is most related to our target, this may due to the fact that the target tag is derived from it. Other tags did not show significant correlation with our target except for the `aggregated features of different contents of tonnage` such as Fines, D50, MB, BPT, CL, ORE and part of the `aggreagted TAIL features`.
- Secondly, we measure the correlations among features, this is because features that have a very high correlation can sometimes be suspicious. 2 columns that have high correlation indicates that they may share a lot of the same information. One of the assumptions of any __parametric predictive model__ is that all features must be independent. From the heatmap, we see that the `tonnage-content related tags`, `the aggregated tonnage-content related tags` are highly correlated within group.
- All in all, the correlations between features and target variable is generally weak. However, to arrive at a conclusive result, more experimentation is needed.

### Existing Feature Engineer Report

- __Existing Feature Engineer Analysis__: 
    - __Quality Models__: the feature engineering process for quality-related model only consider business related factors, for example, the various aggregation of different tonnage contents(SUM/MEAN..). However, other engineered feature type can also be taken into consideration. For example:
        - Feature that can reveal macro patterns that occur over an entire year, such as Ratio/Difference of certain feature among the `latest value and yearly average`. My assumption is that if the recent value of tonnage content, floatation, hydro-transportation and tailing is higher than their yearly average, the quality is likely to be lower than expecation, the machine is likely to flag.
        - Inter-time-period features that may reveal patterns on a micro scale between months, such as features that could `adds more granularity` to the existing feature. My assumption is that the average/maximum temperature, flotation, tailing... across different time periods can be a great indicator of upset pattern. The higher the monthly average/maximum, the more possible that the machine is going to flag. Other features such as the percentage change across different time period can also be strong indicator of upset flagging. My assumption is that if there is a dramatic increase for tonnage content flow rate, the machine is likely to flag. 
    - __Froth-Recovery Models__: the feature engineering process for froth-recovery model contains part of the generation process of quality model such as the tailing and hydro-transport process, besides, it also has general feature engineering process which aggregate using sum/mean/sum-of-change method for all features across prediction window(60min,120min). However,the aggregation method only take the inter-time-period feature generation methods into consideration. More macro-side method could be considered. For example:
        - Feature that can reveal macro patterns that occur over an entire year, such as Ratio/Difference of certain feature among the `latest value and yearly average`. My assumption is that if the recent value of tonnage content, floatation, hydro-transportation and tailing is higher than their yearly average, the froth-recovery rate is likely to be low and thereby likely to flag.
        - Inter-time-period features that may reveal patterns on a micro scale between months. Instead of just aggregating features across hour, we can try a larger time period(such as day/week/month). My assumption is that aggregated features across a larger time period could add more macro patterns to the model.
    - __BASE MODELS__: the four models, HPW, IST, 2 Throughput models share the same feature engineer process which only contain the general feature aggregation process which aggregate using sum/mean/sum-of-change method for all features across prediction window(60min,120min). However, specific engineering process can also be added. For example,
        - For throughput models: we can generate more macro/micro patterns for conveyor related features. My assumption is that if the average conveyor of P85/P82 is lower than expected, the ore throughput is likely to drop(render flagging).
        - For HPW model: more processed features regarding (a) boiler status, (b) heat exchangers, (c) GTG status, (d)PEW properties can be considered. My assumption is that if the maximum temperature across month for heat exchangers drop, the HPW tank temperature is likely to drop(render flagging).
        - For IST model: more feature engineering regarding (a) the ore feed, (b) hydrotransport properties, (c) Sep cell production, (d) tails losses can be considered. My assumption is that if the ore feed reduce across time period, the froth rate is likely to drop and therefore the drop of IST tank(render flagging).
    - __P53 Model__: like base models, it lack the specific feature engineering process. We could add specific engineering process for (a)amine temperature, (b) signs of foaming, (c) steam stripping rate, (d) KO drum level.
